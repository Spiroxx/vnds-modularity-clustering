# 30-Hour Project Roadmap: VNDS Modularity Clustering

This document outlines the step-by-step implementation plan for the 30-hour master's project based on the article *Modularity maximization in networks by variable neighborhood search*[cite: 1].

## Phase 1: Repository Setup & Data Acquisition (Hours 1–4)
*   **Hour 1:** Initialize the version control repository and set up the project environment (e.g., configuring Julia or C++).
*   **Hour 2:** Download the required benchmark datasets to validate the algorithm. Specifically, acquire the `karate` (78 nodes) and `dolphins` (159 nodes) datasets from the 10th DIMACS Implementation Challenge[cite: 1].
*   **Hours 3-4:** Write the parsing scripts to load the graph data from files into memory, ensuring adjacency lists and edge weights are structured efficiently.

## Phase 2: Objective Function & Baseline Heuristic (Hours 5–10)
*   **Hours 5-6:** Implement the evaluation function to calculate the modularity ($Q$) of a given graph partition, translating the weighted modularity equation into code[cite: 1].
*   **Hours 7-10:** Implement the `LPAm+` label propagation heuristic[cite: 1]. Ensure it loops over active labels rather than every single vertex to achieve the near-linear time execution described by the authors[cite: 1].

## Phase 3: The VNS Perturbation Neighborhoods (Hours 11–18)
To escape local maxima, implement the six neighborhood "shaking" routines[cite: 1].
*   **Hours 11-12:** Implement **SINGLETON** (breaking a cluster into isolated nodes) and **DIVISION** (splitting a community randomly in half)[cite: 1].
*   **Hours 13-14:** Implement **NEIGHBOR** (reassigning a vertex to a neighbor's label) and **EDGE** (putting two linked vertices from different clusters into one)[cite: 1].
*   **Hours 15-16:** Implement **FUSION** (merging clusters) and **REDISTRIBUTION** (destroying a cluster and spreading its vertices randomly)[cite: 1].
*   **Hours 17-18:** Write basic unit tests verifying that each perturbation successfully modifies the graph partition without breaking data structures.

## Phase 4: Variable Neighborhood Decomposition Search (Hours 19–26)
*   **Hours 19-22:** Build the decomposition logic (VNDS) to isolate subproblems[cite: 1]. Implement the function that selects a random cluster and its $s-1$ neighboring clusters for local improvement[cite: 1].
*   **Hours 23-26:** Assemble the main optimization loop. Integrate the random selection of the six perturbation operators based on the probability distribution defined in the paper (e.g., 30% Singleton, 28% Neighbor, etc.)[cite: 1].

## Phase 5: Evaluation & Final Polish (Hours 27–30)
*   **Hours 27-28:** Run the completed algorithm against the `karate` and `dolphins` datasets[cite: 1]. Compare your outputs to the exact optimal solutions provided in the paper to guarantee accuracy[cite: 1].
*   **Hours 29-30:** Profile the code to eliminate memory bottlenecks or slow loops. Compile the execution times, modularity scores, and algorithm design into the final project report.
